### Replace Using Mean, Median, or Mode

```python
# using mean() method
# Mean = the average value (the sum of all values divided by number of values).
ddf3 = pd.read_csv('test.csv')
x = ddf3["Height"].mean()
x
```

```python
ddf3["Height"].fillna(x, inplace = True)
ddf3
```

```python
# using median() method
# Median = the value in the middle, after you have sorted all values ascending.
ddf3 = pd.read_csv('test.csv')
x = ddf3["Height"].median()
x
```

```python
ddf3["Height"].fillna(x, inplace = True)
ddf3
```

```python
# using mode() method
# Mode = the value that appears most frequently
ddf3 = pd.read_csv('test.csv')
x = ddf3["Height"].mode()[0]
x
```

```python
ddf3["Height"].fillna(x, inplace = True)
ddf3
```






### Cleaning Wrong Format

```python
# Data set
ddf4 = pd.read_csv('test.csv')
ddf4
```

```python
# Convert all the cells in the "Date" column into dates
ddf4['Age'] = pd.to_datetime(ddf4['Age'])
ddf4
```

```python
# removes rows with NaT: Not a Time
ddf4.dropna(subset=['Age'], inplace = True)
ddf4
```




## Cleaning Wrong Data

```python
# Data set
ddf5 = pd.read_csv('test.csv')
ddf5
```

```python
ddf5.loc[9, 'Height'] = 5.6
ddf5
```

```python
 """
Loop through all values in the "Duration" column.
If the value is higher than 120, set it to 120:
"""
for x in ddf5.index:
    if ddf5.loc[x, "Height"] > 3:
        ddf5.loc[x, "Height"] = 3
ddf5
```

```python


 """
OR, Remove all the rows which have wrong data.
Delete rows where "Height" is lower than 2:
"""
for x in ddf5.index:
    if ddf5.loc[x, "Height"] < 2:
        ddf5.drop(x, inplace = True)
ddf5

# Row 18 will remove
```



## Cleanning Duplicates
Duplicate rows are rows that have been registered more than one time.

```python
ddf6 = pd.read_csv('test.csv')
ddf6
```

```python
# Here Row 19 and 20 are sameq.
# To remove duplicates, use the drop_duplicates() method
ddf6.drop_duplicates(inplace = True)
ddf6
```
